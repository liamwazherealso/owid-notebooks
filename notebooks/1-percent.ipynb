```python
DATA_PATH="../owid-datasets/datasets/Top 1% wealth shares - Chartbook of Economic Inequality (2017)/Top 1% wealth shares - Chartbook of Economic Inequality (2017).csv"
```

```python
df = pd.read_csv(DATA_PATH)
```

```python
df.head()
```

```python
x_min, x_max = df["Year"].min(), df["Year"].max()
```

```python
wealth = "Share of wealth held by top 1% (Chartbook of Economic Inequality 2017)"

y_min, y_max = df[wealth].min(), df[wealth].max()
```

```python
df["Entity"].unique()
```

```python
for e in df["Entity"].unique(): 
    plt.plot(df[df["Entity"] == e]["Year"], df[df["Entity"] == e][wealth], label=e)
plt.axis([x_min, x_max, y_min, y_max ])
plt.legend(loc="upper left")
plt.figure(figsize=(10,20))
plt.show()
```
